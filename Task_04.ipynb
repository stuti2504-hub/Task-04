```python
# Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Create Dataset
data = {
    'Gender': ['Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male'],
    'Parental Education': ['Bachelor', 'High School', 'Master', 'Bachelor', 'Some College', 'Master', 'High School', 'Some College', 'Bachelor', 'Master'],
    'Test Preparation': ['Completed', 'None', 'Completed', 'None', 'Completed', 'None', 'Completed', 'None', 'Completed', 'None'],
    'Study Hours': [4, 2, 5, 3, 4, 2, 6, 3, 5, 2],
    'Math Score': [85, 65, 90, 70, 80, 60, 92, 75, 88, 68],
    'Reading Score': [88, 70, 94, 73, 82, 65, 95, 78, 90, 72],
    'Writing Score': [86, 68, 91, 70, 81, 64, 94, 76, 89, 71]
}

df = pd.DataFrame(data)

# Step 3: Display first few rows
print(df.head())

# Step 4: Descriptive Statistics
print("\nDescriptive Statistics:")
print(df.describe())

# Step 5: Average Scores by Gender
avg_scores = df.groupby('Gender')[['Math Score', 'Reading Score', 'Writing Score']].mean()
print("\nAverage Scores by Gender:")
print(avg_scores)

# Step 6: Visualization 1  Average Scores by Gender
avg_scores.plot(kind='bar', figsize=(8,5))
plt.title("Average Scores by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Score")
plt.legend(title="Subjects")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Step 7: Visualization 2  Relation between Study Hours and Math Score
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Study Hours', y='Math Score', hue='Test Preparation', style='Gender', s=100)
plt.title("Relation between Study Hours and Math Score")
plt.xlabel("Study Hours")
plt.ylabel("Math Score")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# Step 8: Correlation Heatmap
plt.figure(figsize=(6,4))
sns.heatmap(df[['Math Score', 'Reading Score', 'Writing Score', 'Study Hours']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation between Scores and Study Hours")
plt.show()
```
